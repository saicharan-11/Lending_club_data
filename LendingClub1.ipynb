```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y, test_size=0.25,random_state=0)
```

```python
from sklearn.linear_model import LogisticRegression
logs=LogisticRegression(random_state=0)
logs.fit(X_train,y_train)
```

```python
y_pred=logs.predict(X_test)
```

```python
from sklearn.metrics import confusion_matrix
CF=confusion_matrix(y_test,y_pred)
print(CF)
```

```python
from sklearn import metrics
acc=metrics.accuracy_score(y_pred,y_test)
print(acc)
```

```python
from sklearn.svm import SVC
cls=SVC(kernel='rbf', random_state=0)
cls.fit(X_train,y_train)
```

```python
y_pred1=cls.predict(X_test)
```

```python
from sklearn import metrics
acc=metrics.accuracy_score(y_pred1,y_test)
print(acc)
```

```python
from sklearn.ensemble import RandomForestClassifier
cls1=RandomForestClassifier(n_estimators=200, criterion='entropy',random_state=0,oob_score=True)
cls1.fit(X_train,y_train)
```

```python
y_pred2=cls1.predict(X_test)
```
